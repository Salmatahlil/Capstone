```python
# read in data from created csv's
# low_memory does nothing (it's deprecated), but does remove annoying warning message
details_dataframe = pd.read_csv("storm_data_details.csv", low_memory=False)
fatalities_dataframe = pd.read_csv("storm_data_fatalities.csv")
locations_dataframe = pd.read_csv("storm_data_locations.csv")
```

To begin cleaning the data, we decide to filter the data by year. Prior to 1996, the data only tracks weather events of tornados, hail, and thunderstorms. We are interested in the data that also includes other features like droughts, excessive heat, hurricanes, etc. This is present in the data after 1996. Following that, we then decided to drop columns that were not helpful for our analysis or if they contained a lot of null values.

We ended this cell by then cleaning some of the individual columns. The damage columns have multiple instances that equal 0, so I replaced these were necessary. Additionally, when source information was not given, we just assumed it was unknown. Then, finally we dropped the few rows were STATE, STATE_FIPS, and DATA_SOURCE were null which was only 4 rows and cleaned up types where necessary.

```python
# Only data in or after 1996 is useful. Prior data only tracks tornados, hail, and thunderstorms
details_dataframe = details_dataframe[details_dataframe["YEAR"] >= 1996]

# drop columns with a large amount of nulls and that are not important for our analysis
drop_cols = ["Unnamed: 0", "BEGIN_YEARMONTH", "BEGIN_DAY", "BEGIN_TIME", "END_YEARMONTH", "END_DAY", "END_TIME", "WFO",
        "EPISODE_ID", "MAGNITUDE", "MAGNITUDE_TYPE", "FLOOD_CAUSE", "CATEGORY", "TOR_F_SCALE", "TOR_LENGTH", "TOR_WIDTH", 
        "TOR_OTHER_WFO", "TOR_OTHER_CZ_STATE", "TOR_OTHER_CZ_FIPS", "TOR_OTHER_CZ_NAME", "BEGIN_RANGE", "BEGIN_AZIMUTH",
        "BEGIN_LOCATION", "END_RANGE", "END_AZIMUTH", "END_LOCATION", "BEGIN_LAT", "BEGIN_LON", "END_LAT", "END_LON", 
        "EPISODE_NARRATIVE", "EVENT_NARRATIVE"]
details_dataframe.drop(columns=drop_cols, inplace=True)

# if there is no value for damage, assume no damage was done
# also clean up different iterations of 0
details_dataframe.DAMAGE_PROPERTY.replace(["0K", "0.00K", np.nan, ""], 0, inplace=True)
details_dataframe.DAMAGE_CROPS.replace(["0K", "0.00K", np.nan, ""], 0, inplace=True)

# if the Source is null, replace to have Unknown Source
details_dataframe.SOURCE.replace(np.nan, "Unknown Source", inplace=True)

# there is only 1 null state/state_fips remaining, so we can just drop it
details_dataframe.dropna(axis=0, how='any', subset=["STATE", "STATE_FIPS"], inplace=True)

# there are only 3 null in WFO and Data Source, so again we can just drop these rows
details_dataframe.dropna(axis=0, how='any', subset=["DATA_SOURCE"], inplace=True)

# cast state_fips as int, since it is an id essentially
details_dataframe["STATE_FIPS"] = details_dataframe["STATE_FIPS"].astype("int64")
```

The only cleaning remaining was within the damage columns. These columns are read in as strings. Each string is formated with a number in float format followed by a letter. The letter denotes the magnitude of the number:
- K -> Thousand
- M -> Million
- B -> Billion
- T -> Trillion

To handle this we created temporary columns to house the information on magnitudes. Then, we removed the magnitude indicators from the damage column. Then we were able to cast the damage columns into float format and multiple them against the magnitude column to get the true damage values. Finally, we casted these values to ints and dropped the temporary magnitude columns.

Once this was completed we were done cleaning the main storm_data_details dataset.

```python
# set up new columns to hold temporary magnitude values
details_dataframe["damage_property_magnitude"] = 0
details_dataframe["damage_crops_magnitude"] = 0

# clean property damage column. Remove K, M, B, T which represent thousand, million, billion, trillion.
details_dataframe.loc[details_dataframe['DAMAGE_PROPERTY'].str.contains("K", na=False), 'damage_property_magnitude'] = 1000
details_dataframe.loc[details_dataframe['DAMAGE_PROPERTY'].str.contains("M", na=False), 'damage_property_magnitude'] = 1000000
details_dataframe.loc[details_dataframe['DAMAGE_PROPERTY'].str.contains("B", na=False), 'damage_property_magnitude'] = 1000000000
details_dataframe.loc[details_dataframe['DAMAGE_PROPERTY'].str.contains("T", na=False), 'damage_property_magnitude'] = 1000000000000
details_dataframe.DAMAGE_PROPERTY.replace(regex=['K', 'M', 'B', 'T'],value='', inplace=True)
details_dataframe.DAMAGE_PROPERTY.replace("", 0, inplace=True)

# clean crop damage column. Remove K, M, B, T which represent thousand, million, billion, trillion.
details_dataframe.loc[details_dataframe['DAMAGE_CROPS'].str.contains("K", na=False), 'damage_crops_magnitude'] = 1000
details_dataframe.loc[details_dataframe['DAMAGE_CROPS'].str.contains("M", na=False), 'damage_crops_magnitude'] = 1000000
details_dataframe.loc[details_dataframe['DAMAGE_CROPS'].str.contains("B", na=False), 'damage_crops_magnitude'] = 1000000000
details_dataframe.loc[details_dataframe['DAMAGE_CROPS'].str.contains("T", na=False), 'damage_crops_magnitude'] = 1000000000000
details_dataframe.DAMAGE_CROPS.replace(regex=['K', 'M', 'B', 'T'],value='', inplace=True)
details_dataframe.DAMAGE_CROPS.replace("", 0, inplace=True)

# cast them as floats first
details_dataframe["DAMAGE_PROPERTY"] = details_dataframe["DAMAGE_PROPERTY"].astype("float64")
details_dataframe["DAMAGE_CROPS"] = details_dataframe["DAMAGE_CROPS"].astype("float64")

# perform the multiplication between damage and magnitude columns
details_dataframe["DAMAGE_PROPERTY"] = details_dataframe["DAMAGE_PROPERTY"] * details_dataframe["damage_property_magnitude"]
details_dataframe["DAMAGE_CROPS"] = details_dataframe["DAMAGE_CROPS"] * details_dataframe["damage_crops_magnitude"]

# cast the columns as ints. Now damage values are in integer format.
details_dataframe["DAMAGE_PROPERTY"] = details_dataframe["DAMAGE_PROPERTY"].astype("int64")
details_dataframe["DAMAGE_CROPS"] = details_dataframe["DAMAGE_CROPS"].astype("int64")

# magnitude columns no longer needed, so drop them.
details_dataframe.drop(columns=["damage_property_magnitude", "damage_crops_magnitude"], inplace=True)
```

After looking into the other two datasets, the fatalities are already tracked with the data we are using. Therefore, we decided that it would be redundant to join that dataset to the details. So, we then decided to join the locations onto the details (left set) using a left join.

We only added the location column from the locations data and then properly renamed this column to city. Additionally, if no city was specified we decided to replace it with a value that was statewide or multiple. Finally, we removed any duplicate values that were created in the process of the left-join and renamed the columns to appropriate names.

```python
# merge just the city location from the location storm data
result = details_dataframe.merge(locations_dataframe[["EVENT_ID", "LOCATION"]], on="EVENT_ID", how="left")
result.rename(columns={"LOCATION":"CITY"}, inplace=True)

# if there is no city, replace value to be statewide
result.CITY.replace(np.nan, "Statewide/Multiple", inplace=True)

# drop the duplicate rows
result.drop_duplicates(inplace=True)
# drop rows that duplicate everything but the city (some entries will be the same except for the neighboring city)
# we don't neccesarily want this to occur and could cause issues with frequency counts of storms
result.drop_duplicates(subset=result.columns[:-1], inplace=True)

# rename all of the columns to more appropriate and consistent names
result.rename(columns={
    "EVENT_ID":         "Event_id",
    "STATE":            "State",
    "STATE_FIPS":       "State_id",
    "YEAR":             "Year",
    "MONTH_NAME":       "Month",
    "EVENT_TYPE":       "Weather_type",
    "CZ_TYPE":          "County_type",
    "CZ_FIPS":          "County_id",
    "CZ_NAME":          "County",
    "BEGIN_DATE_TIME":  "Start_datetime",
    "CZ_TIMEZONE":      "Timezone",
    "END_DATE_TIME":    "End_datetime",
    "INJURIES_DIRECT":  "Injuries_direct",
    "INJURIES_INDIRECT":"Injuries_indirect",
    "DEATHS_DIRECT":    "Deaths_direct",
    "DEATHS_INDIRECT":  "Deaths_indirect",
    "DAMAGE_PROPERTY":  "Property_damage",
    "DAMAGE_CROPS":     "Crop_damage",
    "SOURCE":           "Reporter_source",
    "DATA_SOURCE":      "Data_source",
    "CITY":             "City"
}, inplace=True)
```

```python
# store all of the changes into a final, clean dataset
result.to_csv("cleaned_storm_data.csv")
```
